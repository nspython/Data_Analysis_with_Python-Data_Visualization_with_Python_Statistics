# Concatenating, Merging, and Joining 

```python
import pandas as pd
```

```python
df1 =pd.DataFrame( {
    'Country' : [ 'USA', 'USA'],
    'Name' : [ 'Sam', 'Jack'],
    'Age' :     [ 20, 22]}, index = [0,1])
```

```python
df2 =pd.DataFrame( {
    'Country' : [ 'CA', 'CA'],
    'Name' : [ 'Elizabeth', 'Tom'],
    'Age' :     [ 23, 21]}, index = [2,3])
```

```python
df3 =pd.DataFrame( {
    'Country' : [ 'GBR', 'GBR'],
    'Name' : [ 'Archie', 'Ruby'],
    'Age' :     [ 32, 31]}, index = [4,5])
```

```python
df1
```

```python
df2
```

```python
df3
```

## Concatenation

```python
# to glue dataframes

pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis = 1)
```

```python
left1 =pd.DataFrame( {'Key' : [ 'K0', 'K1'],
    'Country' : [ 'USA', 'USA'],
    'Name' : [ 'Sam', 'Jack'],
    'Age' :     [ 20, 22]})
```

```python
right1 =pd.DataFrame( {'Key' : [ 'K0', 'K1'],
    'Country' : [ 'CA', 'CA'],
    'Name' : [ 'Elizabeth', 'Tom'],
    'Age' :     [ 23, 21]})
```

```python
left1
```

```python
right1
```

## Merging

```python
pd.merge(left1, right1, how = "inner", on = "Key")
```

```python
left2 =pd.DataFrame( {'Key1' : [ 'K0', 'K0','K1', 'K2'],
    'Key2' : [ 'K0', 'K1','K0', 'K1'],
    'Country' : [ 'USA', 'GBR','CA','TUR'],
    'Name' : [ 'Sam', 'Jack','Jessie','Ali'],
    'Age' :     [ 20, 22, 26, 24]})
```

```python
right2 =pd.DataFrame( {'Key1' : [ 'K0', 'K1','K1', 'K2'],
    'Key2' : [ 'K0', 'K0','K0', 'K0'],
    'Country' : [ 'GER', 'FR','IT','ES'],
    'Name' : [ 'Hans', 'Macron','Sergio','Pablo'],
    'Age' :     [ 23, 21, 28, 25]})
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2, right2, on = ['Key1','Key2']) # default;  how: str = 'inner'
```

```python
pd.merge(left2, right2, how = 'outer', on = ['Key1','Key2'])
```

```python
pd.merge(left2, right2, how = 'right', on = ['Key1','Key2'])
```

```python
pd.merge(left2, right2, how = 'left', on = ['Key1','Key2'])
```

## Joining

```python
left3 =pd.DataFrame( {
    'Country1' : [ 'USA', 'GBR','CA'],
    'Name1' : [ 'Sam', 'Jack','Jessie'],
    'Age1' : [ 20, 22, 26]},
     index = [ 'K0','K1', 'K2'])
```

```python
right3 =pd.DataFrame( {
    'Country2' : [ 'GER','IT','ES'],
    'Name2' : [ 'Hans','Sergio','Pablo'],
    'Age2' :     [ 23, 28, 25]},
    index = [ 'K0','K2', 'K3'])
```

```python
left3
```

```python
right3
```

```python
left3.join(right3)
```

```python
left3.join(right3,how = "outer")
```
